```python
import json
```

```python
with open('/home/pr3ssh/initiatives.json', 'r') as f:
    initiatives = json.loads(f.read())
    print(len(initiatives))
```

```python
for i in initiatives:
    i['oldid'] = str(i['_id'])
    i['_id'] = i['reference'].replace('/', '-')
    if i['initiative_type_alt'] == 'Respuesta':
        i['_id'] += '-respuesta'
```

```python
with open('initiatives-mod.json', 'w') as f:
    f.write(json.dumps(initiatives, ensure_ascii=False, indent=4))
```
